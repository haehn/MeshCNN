```python
%pylab inline
import struct, os
```

```python
train_dir = 'datasets/shrec_16/santa//train/'

for t in os.listdir(train_dir):
    if t.endswith('obj'):
        obj_file = os.path.join(train_dir, t)
        view_meshes(obj_file)
        (vs, faces, edges), scale_by  = parse_obje(obj_file,1)
        
        compressed = TrakoDracoPy.encode_mesh_to_buffer(vs.ravel(), faces.ravel())
        pixels = np.array(struct.unpack('B'*len(compressed[5:]), compressed[5:]))
        new_size = (int(sqrt(pixels.shape[0]))+1, int(sqrt(pixels.shape[0]))+1)
        new_pixels = np.zeros(new_size[0]*new_size[1])
        new_pixels[0:len(pixels)] = pixels
        new_pixels = new_pixels.reshape(new_size[0], new_size[1])
        print(new_size)
        plt.imshow(new_pixels)
        
```
